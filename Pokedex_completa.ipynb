```python
# pegando todas as descrições, html das imagens e o nome* dos pokemons do site https://sg.portal-pokemon.com/. *pegando o nome para comparar quais pokemons faltam
versao1 = []
versao2 = []
htmlImg = []
namePoke = []
for i in range(1,1026):
  f = 0
  url = f'https://sg.portal-pokemon.com/play/pokedex/{str(i).zfill(4)}'
  reqs = requests.get(url)



  soup = BeautifulSoup(reqs.text, 'html.parser')
  #HTML da imagem
  imagem = soup.find('img', class_ = 'pokemon-img__front').get('src')
  htmlImg.append(f'https://sg.portal-pokemon.com{imagem}')
  #Nome do pokemon
  pokeName =  soup.find('p',class_ = 'pokemon-slider__main-name').text
  namePoke.append(pokeName)
  descricion = soup.find_all('p',class_ = "pokemon-story__body size-14")
  for item in descricion:
      if f == 0:
        versao1.append(item.get_text())

      if f == 1:
        versao2.append(item.get_text())
      f+=1
  #verificar se existe megaevoluções
  for h in range(1,6):
    if requests.get(f'https://sg.portal-pokemon.com/play/pokedex/{str(i).zfill(4)}_{h}').status_code == 200:
      soup = BeautifulSoup(requests.get(f'https://sg.portal-pokemon.com/play/pokedex/{str(i).zfill(4)}_{h}').text, 'html.parser')

      #HTML da imagem
      imagem = soup.find('img', class_ = 'pokemon-img__front').get('src')
      htmlImg.append(f'https://sg.portal-pokemon.com{imagem}')
      #Nome do Pokemon, alguns pokemons possuem um subname como o Zygard 10% Forme, por isso a concatenação, caso não tenha subname, o strip() no append resolve o espaço
      pokeName =  soup.find('p',class_ = 'pokemon-slider__main-name',).text + " " + soup.find('p',class_ = 'pokemon-slider__main-subname',).text
      namePoke.append(pokeName.strip())
      f = 0
      descricion = soup.find_all('p',class_ = "pokemon-story__body size-14")
      for item in descricion:
        if f == 0:
          versao1.append(item.get_text())

        if f == 1:
          versao2.append(item.get_text())
        f+=1



```

```python
desc1 = versao1
desc2 = versao2
descricao1_normalizada = []
descricao2_normalizada = []
for item in desc1:
  descricao1_normalizada.append(item.split('\n')[1])
for item in desc2:
  descricao2_normalizada.append(item.split('\n')[1])
```
